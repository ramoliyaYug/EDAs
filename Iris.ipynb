```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8-pastel")

sns.set_theme(
    style="whitegrid",
    palette="pastel",
    context="notebook"
)
```

```python
df = sns.load_dataset("iris")
```

```python
numeric_colums = df.select_dtypes(include='number')
categorical_colums = df.select_dtypes(include='object')
```

```python
corelation_matrix = numeric_colums.corr()
print(corelation_matrix)
```

```python
df.hist(figsize=(10,10))
plt.show()
```

```python
sns.kdeplot(df['sepal_length'], fill=True)
plt.show()
```

```python
species_count = df['species'].value_counts()
print(species_count)
```

```python
sns.countplot(data=df,x='species')
plt.show()
```

```python
sns.scatterplot(data=df,x='sepal_length',y='sepal_width',hue='species')
plt.title('Sepal Length Vs Sepal Width')
plt.show()
```

```python
sns.pairplot(df, hue='species')
plt.show()
```

```python
for cols in numeric_colums:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='species', y=cols, data=df)
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(corelation_matrix, annot=True, cmap="coolwarm")
plt.show()
```
